## Need to sort out not all labels appearing on the new dataframe

```python
new_df = pd.DataFrame(columns=['url', 'repo', 'title', 'body', 'num_labels', 'labels', 'c_bug', 'c_feature', 'c_question', 'class_int'])
```

```python
new_data = []

for index, row in subset_df.iterrows():
    if 'issue' in row['payload']:
        url = row['payload']['issue']['html_url']
        repo = row['repo']['name']
        title = row['payload']['issue']['title']
        body = row['payload']['issue']['body']
        num_labels = len(row['payload']['issue']['labels'])
        labels = [label['name'] for label in row['payload']['issue']['labels']]
        
        # Define class_int based on label presence
        if 'bug' in labels:
            class_int = 0
        elif any(label in ['feature', 'enhancement', 'improvement', 'request'] for label in labels):
            class_int = 1
        elif any(label in ['question', 'discussion'] for label in labels):
            class_int = 2
        else:
            class_int = -1  # Handle cases where no class label is found
        
        # Append the extracted data as a dictionary
        new_data.append({
            'url': url,
            'repo': repo,
            'title': title,
            'body': body,
            'num_labels': num_labels,
            'labels': labels,  # Store all labels for each issue as a list
            'class_int': class_int
        })

# Create the new DataFrame using pandas.concat
new_df = pd.DataFrame(new_data)
```

## Setting train and test data

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
import torch
from torch.utils.data import DataLoader, Dataset
from torch.nn import functional as F
from sklearn.metrics import classification_report
```

```python
final_subset = new_df[new_df['class_int'] != -1]
```

```python
final_subset.groupby('class_int').size()
```

```python
from sklearn.model_selection import train_test_split
```

```python
final_subset['text'] = final_subset['title'] + ' ' + final_subset['body']

# Encode the 'class_int' labels
label_encoder = LabelEncoder()
final_subset['class_int'] = label_encoder.fit_transform(final_subset['class_int'])

# Split the data into training and testing sets
train_df, test_df = train_test_split(final_subset, test_size=0.2, random_state=42)
```

```python
final_subset
```

# Data and Text pre-processing

```python
# Define a custom dataset
class TextClassificationDataset(Dataset):
    def __init__(self, text, labels, tokenizer, max_length):
        self.text = text
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.text)

    def __getitem__(self, idx):
        text = str(self.text[idx])
        labels = self.labels[idx]

        encoding = self.tokenizer(
            text,
            add_special_tokens=True,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt',
            return_token_type_ids=False,
            return_attention_mask=True,
        )

        return {
            'input_ids': encoding['input_ids'].squeeze(),
            'attention_mask': encoding['attention_mask'].squeeze(),
            'labels': torch.tensor(labels, dtype=torch.long),
        }
```

```python
# Load a pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3)
```

```python
# Create DataLoader instances for training and testing
max_length = 256  # You can adjust this based on your dataset and resource limitations
train_dataset = TextClassificationDataset(
    train_df['text'].values,
    train_df['class_int'].values,
    tokenizer,
    max_length,
)
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)

test_dataset = TextClassificationDataset(
    test_df['text'].values,
    test_df['class_int'].values,
    tokenizer,
    max_length,
)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)
```

```python
# Define training parameters
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
optimizer = AdamW(model.parameters(), lr=2e-5)
num_epochs = 3  # You can adjust this
model.train()
```

```python
# Training loop
for epoch in range(num_epochs):
    for batch in train_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
```

```python
# Evaluation
model.eval()
predictions = []
true_labels = []

with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].cpu().numpy()

        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        predicted_labels = np.argmax(logits.cpu().numpy(), axis=1)

        predictions.extend(predicted_labels)
        true_labels.extend(labels)
```

```python
# Decode predicted labels back to original class_int values
predicted_labels = label_encoder.inverse_transform(predictions)
```

```python
# Generate a classification report
report = classification_report(true_labels, predicted_labels)
print(report)
```
